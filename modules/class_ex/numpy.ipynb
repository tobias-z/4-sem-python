```python
import numpy as np
```

```python
# Table : find the slicing
a: np.ndarray = np.arange(10, 30).reshape(4, 5)
print(a)

yellow = a[0][0]
blue = a[:,1::2]
green = a[0:3, 2]
purple = a[::2, -1]

print(f"yellow: {yellow}\n blue: {blue}\n green: {green}\n purple: {purple}")
```

```python
# 
a: np.ndarray = np.arange(0, 27).reshape((3, 3, 3))
print(a[1, 1, :])
print(a[:, 1, 0])
print(a[0, :, 2])

print(a)
```

```python
a = np.linspace(0, 2 * np.pi, 50)
print(a)
print("_____________________________________________________________________________\n")
print(np.sin(a))
```

```python
# Masking
data: np.ndarray = np.arange(1, 101).reshape(10, 10)
print(data)
even_number_mask = data % 2 == 0

even = data[even_number_mask]
print(even)

copy = np.copy(even)
ends_with = np.where(print(str(copy)[-1] == "6"))
print(even[ends_with])
```
