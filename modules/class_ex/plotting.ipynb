```python
student_attendance = {'day1':33, 'day2':34,'day3':29,'day4':31,'day5':28,'day6':26,'day7':30}
list(student_attendance.keys())
```

```python
import matplotlib.pyplot as plot

keys = list(student_attendance.keys())
values = list(student_attendance.values())

plot.plot(keys, values)
plot.title("Student Attendance", fontsize=24)
plot.xlabel("Days", fontsize=14)
plot.ylabel("Students", fontsize=14)
```
